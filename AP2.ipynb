```python
def add(a, b):  # Code khodam
    return a + b

def product(a, b):
    return a * b

def division(a, b):
    return a / b

def subtract(a, b):
    return a - b

# سوئیچ برای انتخاب تابع مناسب
def calculator(a, b, operator):
    operations = {
        'add': add,
        'product': product,
        'division': division,
        'subtract': subtract
    }
    func = operations.get(operator, None)
    if func is None:
        return "Invalid operation"
    return func(a, b)


# --- گرفتن ورودی ---
a, b = map(int, input("Enter two numbers: ").split())
operator = input("Enter your operation (add, product, division, subtract): ")

print(calculator(a, b, operator))
```

```python
def average(key='ali', value='9', **kwargs):
   """ Collect all parameters into a single dictionary
   """
   result = {key: value}
   result.update(kwargs)
   return result
all_results = {} 

# Code khodam

while True:
    user_input = input("Enter name and average (or type 'escape' to exit): ")

    if user_input.lower() == "escape":
        break

    key, value = user_input.split()

    result = average(**{key: value})
    all_results.update(result)

    print(all_results)
```
