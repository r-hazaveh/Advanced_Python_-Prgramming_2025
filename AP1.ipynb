```python
def filter_by_id(data1,userId):
    return list(x['title'] for x in data1["posts"] if x['userId'] == userId)

filter_by_id(data1,3)   # List Comprehension
```

```python


def fib(n):
    if n == 0 or n == 1:
        return 1
    return fib(n-1) + fib(n-2)
    
# def count_fib(p):
#     lst = []
#     for i in range(p):
#        lst.append(fib(i))
#     return lst

def count_fib(p):
    return [fib(i) for i in range(p)]
```
