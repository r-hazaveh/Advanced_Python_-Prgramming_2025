```python
import os
import json
import uuid


# ============================
#       SaveData  (Mixin)
# ============================
class SaveData:

    def save(self):

        file_path = self.__class__._data_file_address   #  ⭐ تغییر: استفاده از protected

        # اگر فایل وجود ندارد → ایجاد لیست خالی
        if not os.path.exists(file_path):
            with open(file_path, 'w', encoding='utf-8') as fp:
                json.dump([], fp)

        # خواندن فایل
        with open(file_path, 'r', encoding='utf-8') as fp:
            try:
                data = json.load(fp)
            except json.JSONDecodeError:
                data = []

        # Update
        updated = False
        for item in data:
            if item['id'] == self.id:       # ⭐ id اکنون property است
                # مقداردهی مجدد
                item.update(self.to_dict())
                updated = True
                break

        # Insert
        if not updated:
            data.append(self.to_dict())

        # Save back
        with open(self.__class__._data_file_address, 'w', encoding='utf-8') as fp:
            json.dump(data, fp, indent=4)



# ============================
#       LoadData (Mixin)
# ============================
class LoadData:

    @classmethod
    def load(cls, id):
        file_path = cls._data_file_address   # ⭐ تغییر

        with open(file_path, "r", encoding="utf-8") as fp:
            data = json.load(fp)

        for item in data:
            if item["id"] == id:
                return cls(**item)     # ⭐ مقدار id خصوصی در __init__ ست می‌شود

        raise ValueError("Invalid ID")



# ============================
#       DeleteData (Mixin)
# ============================
class DeleteData:

    def delete(self, id):

        file_path = self.__class__._data_file_address   # ⭐ تغییر

        if not os.path.exists(file_path):
            print("There is no data to delete")
            return False

        # خواندن فایل
        with open(file_path, 'r', encoding='utf-8') as fp:
            try:
                data = json.load(fp)
            except json.JSONDecodeError:
                print("There is no data to delete")
                return False

        # حذف رکورد
        new_data = [item for item in data if item['id'] != id]

        if len(new_data) == len(data):
            print("No record found with this ID.")
            return False

        # نوشتن داده جدید
        with open(self.__class__._data_file_address, 'w', encoding='utf-8') as fp:
            json.dump(new_data, fp, indent=4)

        return True



# ============================
#       Address Class
# ============================
class Address(SaveData, DeleteData, Test, LoadData):

    _data_file_address = "./data1/address.json"   # ⭐ تغییر → protected

    def __init__(self, street, alley, postal_code, id=None):
        self.street = street
        self.alley = alley
        self.postal_code = postal_code

        # ⭐ id → private
        if id is not None:
            self.__id = id     # ⭐ private
        else:
            self.__id = str(uuid.uuid4())   # ⭐ private

    # ⭐ property getter
    @property
    def id(self):
        return self.__id
    
    # def get_id(self):  # برای خواندن اتریبوت خصوصی getter
    #     return self.__id
    

    # def set_id(self, value): #  گذاشتن شرط روی اتریبیوتهای خصوصی , مقداردهی اتربیوتهای خصوصیsetter   
    #    value = str(value)   # ⭐ تبدیل به رشته
    
    #    if value.isdigit():  # ⭐ فقط عدد بودن
    #         self.__id = value
    #    else:
    #         raise ValueError("ID must contain only digits")


    @id.setter  #Setter
    def id(self, value): #  گذاشتن شرط روی اتریبیوتهای خصوصی , مقداردهی اتربیوتهای خصوصیsetter   
       value = str(value)   # ⭐ تبدیل به رشته
    
       if value.isdigit():  # ⭐ فقط عدد بودن
            self.__id = value
       else:
            raise ValueError("ID must contain only digits")



    # ⭐ خروجی قابل ذخیره‌سازی
    def to_dict(self):
        return {
            "id": self.__id,
            "street": self.street,
            "alley": self.alley,
            "postal_code": self.postal_code
        }

    def __str__(self):
        return f"Address({self.street}, {self.alley}, {self.postal_code}) - ID: {self.__id}"
```

```python
# ایجاد وررودی برای دکوراتور
# دکوراتورها برروی تابع در واقع کار ارث بری را برروی کد انجام میدهند.
# در واقع دکوراتور ورودی یک دکوراتور کلی تر میشود
```

```python
import functools # Decorators With Arguments

def repeat(num_times):
    def decorator_repeat(func):
        @functools.wraps(func)  # همیشه هست 
        def wrapper_repeat(*args, **kwargs):
            for _ in range(num_times):
                value = func(*args, **kwargs)
            return value
        return wrapper_repeat
    return decorator_repeat
```

```python
@repeat(num_times=5)
def greet(name):
    print(f"Hello {name}")


greet("World")
```

```python
import time
import functools

def calculate_time(func):               # func همان تابع اصلی است
    @functools.wraps(func)              # حفظ نام، داک‌استرینگ و ...
    def wrapper(*args, **kwargs):       # پذیرفتن تمامی ورودی‌های تابع
        start = time.time()             # شروع زمان
        result = func(*args, **kwargs)  # اجرای تابع اصلی
        end = time.time()               # پایان زمان
        
        print("Execution Time:", end - start)
        return result                   # بازگشت خروجی تابع
    return wrapper                      # بازگشت تابع wrapper

@calculate_time   # میتوان چنددکوراتور را روی تابع اصلی اجرا کرد
def calculator2(n):  # دکوراتوری که به تابع اصلی نزدیکتر است زودتر اجرا میشود 
    s = 0
    for i in range(1, n):
        s += 1 / (i ** (1/3))
    return s
```

```python
import time
import functools  # Decorators With Arguments


def calculate_time(repeat):         # 1️⃣ دکوراتور ورودی می‌گیرد
    def real_decorator(func):       # 2️⃣ تابع اصلی را می‌گیرد
        @functools.wraps(func)
        def wrapper(*args, **kwargs):    # 3️⃣ اجراکننده
            start = time.time()
            result = func(*args, **kwargs)
            end = time.time()

            for _ in range(repeat):
                print("Execution time:", end - start)

            return result
        return wrapper
    return real_decorator
```

```python
@calculate_time(3)
def add(a, b):
    return a + b

print(add(5, 10))
```

```python
import time
import functools  # Nested Decorators

def timer(func):
    @functools.wraps(func)
    def wrapper(*args, **kwargs):
        start = time.time()
        result = func(*args, **kwargs)
        end =  time.time()
        print("Time:", end - start)
        return result
    return wrapper

def logger(func):
    @functools.wraps(func)
    def wrapper(*args, **kwargs):
        print(f"Calling {func.__name__}")
        return func(*args, **kwargs)
    return wrapper

@timer
@logger
def compute(n):
    return sum(i*i for i in range(n))

compute(100000)
```

```python
def my_decorator(func):
    def wrapper(*args, **kwargs):
        print("Before method runs")
        result = func(*args, **kwargs)
        print("After method runs")
        return result
    return wrapper

class Person:
    
    @my_decorator
    def say_hello(self):
        print("Hello!")

p = Person()
p.say_hello()
```
